### Creating the target column (time taken for delivery) from order timestamp and delivery timestamp

```python
df['delivery_time']=df['actual_delivery_time']-df['created_at']
df['delivery_time_sec']=df['delivery_time'].dt.total_seconds()
df['delivery_time_min']=df['delivery_time_sec']//60
```

### Remove outliers

```python
Q1=df_num.quantile(0.25)
Q3=df_num.quantile(0.75)
IQR=Q3-Q1
df1=df_num[~((df_num<(Q1-1.5*IQR))|(df_num>(Q3+1.5*IQR))).any(axis=1)]
df1.head()
```

### Encoding categorical column

```python
data=pd.get_dummies(data,columns=['order_day'],dtype=int)
data
```

```python
data=data.drop(['created_at','actual_delivery_time'],axis=1)
```

### Data scaling

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scale_df=scaler.fit_transform(data)
data1=pd.DataFrame(scale_df,columns=data.columns)
data1
```

### Training a random forest model

```python
from sklearn.model_selection import train_test_split
x=data1
y=data1['delivery_time_min']
x=x.drop('delivery_time_min',axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
MSE=mean_squared_error(y_test,y_pred)
R2_score=r2_score(y_test,y_pred)
print(f"Mean Squred Error : {MSE}")
print(f"R2 Score :{R2_score}")
```

### Defining NN architecture

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train,X_val,Y_train,Y_val=train_test_split(x,y,test_size=0.30)
```

```python
X_test,X_val,Y_test,Y_val=train_test_split(X_val,Y_val,test_size=0.30)
```

```python
print('Train : ', X_train.shape, y_train.shape)
print('Validation:', X_val.shape, Y_val.shape)
print('Test  : ', X_test.shape, Y_test.shape)
```

```python
def baseline():
   model = Sequential([
                    Dense(256, activation="linear"),
                    Dense(128, activation="linear"),
                    Dense(64, activation="linear"),
                    Dense(1)])
   return model
```

```python
model=baseline()
```

```python
model.compile(optimizer = tf.keras.optimizers.Adam(),
                loss = tf.keras.losses.MeanSquaredError(),
                metrics=["R2Score"])
```

```python
Model = model.fit(X_train, Y_train, validation_data = (X_val, Y_val),  epochs=10, batch_size=128, verbose=0)
```

```python
model.evaluate(X_train,Y_train)
```

```python
model.evaluate(X_val,Y_val)
```
